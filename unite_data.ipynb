```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from catboost import CatBoostClassifier
```

```python
train = pd.read_csv('data/train.csv')
train_agg = pd.read_feather('new_data/train_agg_2.ftr')
train_geo = pd.read_csv('new_data/train_geo_2.csv')
```

```python
train = train.sort_values('field_id').reset_index(drop=True)
train_geo = train_geo.sort_values('field_id').reset_index(drop=True)
train_agg = train_agg.sort_values('field_id').reset_index(drop=True)
```

```python
train_all = pd.DataFrame()

for col in train_geo.columns:
    train_all[col] = train_geo[col]
    
for col in train_agg.columns[:-1]:
    if 'B' not in col:
        train_all[col] = train_agg[col]
    
for col in train.columns[2:2130]:
    train_all[col] = train[col]
```

```python
train_all = train_all.reset_index(drop=True)
```
