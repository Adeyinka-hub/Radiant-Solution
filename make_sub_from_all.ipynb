# FINAL SUB

```python
sub = pd.read_csv('sub_all_all.csv')
cols = sub.columns[1:]
```

```python
from glob import glob
```

```python
all_subs = []

for p in glob('subs/sub*.csv'):
    all_subs.append(pd.read_csv(p)[cols].values)
    
for p in glob('lama_sub/sub*.csv'):
    all_subs.append(pd.read_csv(p)[cols].values)
    
for p in glob('new_subs/sub*.csv'):
    all_subs.append(pd.read_csv(p)[cols].values)
```

```python
sub[cols] = np.median(all_subs, axis=0)
```

# Difference btw model with geo and model without geo

```python
nogeo = pd.read_csv('lama_sub/sub_lama_nogeo.csv')
geo = pd.read_csv('lama_sub/sub_lama_2.csv')
```

```python
geo['max_prob'] = geo[cols].apply(lambda x: x.max(), axis=1)
nogeo['max_prob'] = nogeo[cols].apply(lambda x: x.max(), axis=1)

geo['max_ind'] = geo[cols].apply(lambda x: np.where(x == x.max())[0][0], axis=1)
nogeo['max_ind'] = nogeo[cols].apply(lambda x: np.where(x == x.max())[0][0], axis=1)

diff = pd.DataFrame()

diff['geo_max_prob'] = geo['max_prob']
diff['nogeo_max_prob'] = nogeo['max_prob']

diff['geo_max_ind'] = geo['max_ind']
diff['nogeo_max_ind'] = nogeo['max_ind']
```
